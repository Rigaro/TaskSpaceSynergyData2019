# Plots

## Hand path difference

### Over iterations

```python
# Config figure
f = plt.figure(figsize=(15,15))
gs = f.add_gridspec(2, 2)
sns.set_context("poster", font_scale=1.5, rc={"lines.linewidth": 5.0, "axes.edgecolor": "0", "grid.color": ".5", "palette":"bright"})
sns.set_style("whitegrid", {"font_scale":"1", "axes.edgecolor": "0", "grid.color": ".5"})
#sns.set_palette()
#sns.set_context("poster", font_scale=1.5)
#sns.set(font_scale=3)

# Plot data
ax = f.add_subplot(gs[0, 0])
sns.lineplot(x="iteration", y="hdiff", hue="case_nm", style="case_nm", data=hdiff[hdiff["target"]=="Close"], hue_order=["", "EP", "TS", "JS"], style_order=["", "EP", "TS", "JS"], estimator=np.median)
# Configure axes
ax.set_title("Target: Close")
ax.set_xlabel('')
ax.set_ylabel('')
plt.xlim(1, 30)
plt.ylim(0, 200)
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles=handles[1:], labels=labels[1:]).set_title("")

ax = f.add_subplot(gs[0, 1])
sns.lineplot(x="iteration", y="hdiff", hue="case_nm", style="case_nm", data=hdiff[hdiff["target"]=="Mid"], hue_order=["", "EP", "TS", "JS"], style_order=["", "EP", "TS", "JS"], estimator=np.median)
# Configure axes
ax.set_title("Target: Mid")
ax.set_xlabel('')
ax.set_ylabel('')
plt.xlim(1, 30)
plt.ylim(0, 200)
ax.get_legend().remove()
#handles, labels = ax.get_legend_handles_labels()
#ax.legend(handles=handles[1:], labels=labels[1:]).set_title("")

ax = f.add_subplot(gs[1, 0])
sns.lineplot(x="iteration", y="hdiff", hue="case_nm", style="case_nm", data=hdiff[hdiff["target"]=="Far"], hue_order=["", "EP", "TS", "JS"], style_order=["", "EP", "TS", "JS"], estimator=np.median)
# Configure axes
ax.set_title("Target: Far")
ax.set_xlabel('Iteration')
ax.set_ylabel('')
plt.xlim(1, 30)
plt.ylim(0, 200)
ax.get_legend().remove()
#handles, labels = ax.get_legend_handles_labels()
#ax.legend(handles=handles[1:], labels=labels[1:]).set_title("")

ax = f.add_subplot(gs[1, 1])
sns.lineplot(x="iteration", y="hdiff", hue="case_nm", style="case_nm", data=hdiff[hdiff["target"]=="High"], hue_order=["", "EP", "TS", "JS"], style_order=["", "EP", "TS", "JS"], estimator=np.median)
# Configure axes
ax.set_title("Target: High")
ax.set_xlabel('Iteration')
ax.set_ylabel('')
plt.xlim(1, 30)
plt.ylim(0, 200)
ax.get_legend().remove()
#handles, labels = ax.get_legend_handles_labels()
#ax.legend(handles=handles[1:], labels=labels[1:]).set_title("")

f.text(-0.01, 0.5, 'Total difference from AB path (cm)', va='center', rotation='vertical')
f.tight_layout()

# Save figure
plt.savefig(dataset + "/hdiff.pdf", format="pdf", bbox_inches = 'tight')
```
